```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("samples/01-se-station.jpg")
```

```python
median = np.median(img)
sigma = 0.33
lower = int(max(0, (1.0 - sigma) * median))
higher = int(min(255, (1.0 + sigma) * median))
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(gray, lower, higher)
```

```python
plt.imshow(edges)
```

```python
cv2.imwrite('edges.jpg',edges)
```

```python
lines = cv2.HoughLines(edges,1,np.pi/180,300)
for l in lines[:50]:
    print(l)
    for rho, theta in l:
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a*rho
        y0 = b*rho
        x1 = int(x0 + 10000*(-b))
        y1 = int(y0 + 10000*(a))
        x2 = int(x0 - 10000*(-b))
        y2 = int(y0 - 10000*(a))

        cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
```

```python
cv2.imwrite('houghlines3.jpg',img)
```
