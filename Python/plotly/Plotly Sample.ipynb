```python
from random import gauss, randint
from pandas import DataFrame as df
```

```python
def simple_random_df(length=1000, min_=0, max_=100):
    """
    Generates a dataframe containing `length` lines of
    integers randomly selected from `min_` to `max_`
    """
    d = df([randint(min_, max_) for _ in range(length)])
    return d

def gauss_random_df(length=1000, mean=50, stdev=2):
    """
    Generates a dataframe with `length` lines of
    numbers drawn from a gaussian with given
    `mean` and `stdev`
    """
    d = df(gauss(mean, stdev) for _ in range(length))
    return d
```

```python
# Standard plotly imports
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode

# Using plotly + cufflinks in offline mode
import cufflinks
cufflinks.go_offline(connected=True)
init_notebook_mode(connected=True)
```

```python
data = gauss_random_df(10000)

data[0].iplot(kind="hist", xTitle="Cell number", yTitle="A random number",
              title="A very (pseudo) random histogram")
```

```python
data[1] = data.applymap(lambda x: x * 2)
data[[0, 1]].iplot(kind="hist")
```
